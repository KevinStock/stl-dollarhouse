```python
import pandas as pd
import gmaps
import json
import requests
from config import gkey
```

```python
# import properties from csv
properties = pd.read_csv("3-1-19-DHPP-List.csv")
properties.head()
```

```python
# geocode addresses
base_url = "https://maps.googleapis.com/maps/api/geocode/json?key=" + gkey + "&address="

lat = []
lng = []

for index, row in properties.iterrows():
    url = base_url + row["Address"].replace(" ", "%20") + ",Saint%20Louis,MO"
    geo_data = requests.get(url).json()
    lat.append(geo_data["results"][0]["geometry"]["location"]["lat"])
    lng.append(geo_data["results"][0]["geometry"]["location"]["lng"])
    print("Process " + row["Address"])

properties["Latitude"] = lat
properties["Longitude"] = lng
```

```python
#properties.to_csv("properties.csv", index=False)
properties = pd.read_csv("properties.csv")
# setup map figure
gmaps.configure(api_key=gkey)
figure_layout = {
    'width': '800px',
    'height': '600px',
    'border': '1px solid black',
    'padding': '1px',
    'margin': '0 auto 0 auto'
}
fig = gmaps.figure(layout=figure_layout)
```

```python
# get marker coordinates
coordinates = []
for index, row in properties.iterrows():
    coordinates.append(tuple((row["Latitude"], row["Longitude"])))

markers = gmaps.marker_layer(coordinates)
fig.add_layer(markers)
fig
```
